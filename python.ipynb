List Comprehension

```python
#T·∫°o list c√°c s·ªë chia h·∫øt cho 3 trong 0‚Äì50.

a = [x for x in range(50) if x % 3 == 0]
print(a)
```

```python
# T·∫°o list ch·ª©a c√°c ch·ªØ c√°i vi·∫øt hoa t·ª´ chu·ªói "data engineer".
txt = "Data engineer"
b = [x for x in txt if x.upper() == x and x != ' ']
print(b)
```

```python
# T·∫°o list c√°c t√≠ch (i * j) cho i, j trong 1‚Äì3.
c = [ i * j for i in range(1,4) for j in range (1,4)]
print(c)
```

Dict Comprehension

```python
# T·ª´ list ['a', 'b', 'c'] v√† [1, 2, 3], t·∫°o dict t∆∞∆°ng ·ª©ng { 'a':1, 'b':2, 'c':3 }.
char = ['a','b','c']
num = [1,2,3]

d = {c : n for c,n in zip(char,num)}
print(d)
```

```python
# ƒê·∫£o ng∆∞·ª£c key/value trong m·ªôt dict.
e = {'a': 1, 'b': 2, 'c': 3}

f = {b : a for a,b in e.items()}
print(f)
```

```python
 #T·∫°o dict ch·ªâ g·ªìm c√°c key c√≥ value ch·∫µn t·ª´ dict ban ƒë·∫ßu.
g = {'a': 1, 'b': 2, 'c': 3}
h = {a:b for a,b in g.items() if b % 2 == 0}
print(h)
```

enumerate()

```python
names = ["Alice", "Bob", "Charlie"]
for idx, name in enumerate(names):
    print(idx, name)
```

```python
#Vi·∫øt h√†m find_indices(lst, value) tr·∫£ v·ªÅ danh s√°ch index c·ªßa value trong lst (d√πng enumerate).

def find_indices(lst,value):
    return [idx for idx,name in enumerate(lst) if name == value]

lst = [1,2,3,2]
value = 2
find_indices(lst,value)
```

4. zip()

```python
names = ["A", "B", "C"]
scores = [85, 90, 78]
for name, score in zip(names, scores):
    print(name, score)
```

```python
# T√≠nh t·ªïng t·ª´ng c·∫∑p ph·∫ßn t·ª≠ c·ªßa hai list c√πng ƒë·ªô d√†i.
names = [1,2,3]
scores = [85, 90, 78]
i = [a+b for a,b in zip(names,scores)]
print(i)
```

```python
#Vi·∫øt h√†m nh·∫≠n tuple (name, age, city) v√† unpack tr·ª±c ti·∫øp trong for.
people = [
    ("Alice", 25, "Hanoi"),
    ("Bob", 30, "Danang"),
    ("Charlie", 28, "HCM")
]


def show_people_info(people: list[tuple[str, int, str]]) -> None:
    """
    In ra th√¥ng tin t·ª´ng ng∆∞·ªùi t·ª´ danh s√°ch tuple (name, age, city)
    """
    for name, age, city in people:
        print(f"{name} is {age} years old and lives in {city}")


show_people_info(people)
```
